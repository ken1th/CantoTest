### I changed the columns names since there were columns with same names. Also, in this analysis we only considered the score, so I filtered the columns accordingly.

```r
names(cantoneseTestRaw) <- make.unique(names(cantoneseTestRaw), sep = "_")
cantoneseTestScore <- cantoneseTestRaw %>% select(contains("[分數]"))
head(cantoneseTestScore)
```

### Hovever, the column names were still not convenient enough for the analysis, so I extracted them and assigned the QTAG code to each question

```r
qtag <- rio::import("QuestionTag.csv")
```

```r
names(cantoneseTestScore) <- qtag[, 2]
head(cantoneseTestScore)
```

```r
#the last two questions were irrelevant
cantoneseTestScore <- cantoneseTestScore %>% select(- Q118, - Q119)
```

### The next step was to convert the fractions into the decimals. 1 would mean the participant answered the question correctly.

```r
TestScoreNumeric <- lapply(colnames(cantoneseTestScore), function(x) {
    temDf <- cantoneseTestScore %>% select(all_of(x)) %>% 
                separate(x, 
                         c("Score", "Full"), 
                         sep = "/") %>% suppressWarnings %>% 
                mutate(Score1 = as.numeric(Score) / as.numeric(Full))
    names(temDf)[names(temDf) == "Score1"] <- x
    return(temDf %>% select(-Score, -Full))
}) %>% bind_cols
head(TestScoreNumeric)
```

### We can now find the correct proportion for each question

```r
NumOfCorrect <- TestScoreNumeric %>% summarise(across(everything(), function(x) sum(x == 1, na.rm = T)))
NumOfAnswer <- TestScoreNumeric %>% summarise(across(everything(), function(x) sum(!is.na(x))))                                                     

CorrectProportionDf <- data.frame(NumOfCorrect = t(NumOfCorrect), NumOfAnswer = t(NumOfAnswer))

CorrectProportionDf <- tibble::rownames_to_column(CorrectProportionDf, "QTAG") %>% 
                    mutate(CorrectProportion = NumOfCorrect / NumOfAnswer) %>% 
                    left_join(qtag, by = "QTAG")
CorrectProportionDf %>% arrange(CorrectProportion)
```

```r
#output the above table to a csv file
write.csv(CorrectProportionDf, "CantoTestCorrectProportion.csv", row.names = F)
```
